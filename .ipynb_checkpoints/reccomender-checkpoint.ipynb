```python
import numpy as np
import pandas as pd
```

```python
anime = pd.read_csv('animes.csv')
```

```python
anime.head()
```

```python
anime.shape
```

```python
anime.head(1)
```

```python
# Cols that are needed for reccomendation system 
# Genre
# id
# synopsis 
# title 

anime = anime[['uid', 'title', 'synopsis', 'genre']]
```

```python
anime.info()
```

```python
# This is the dataframe where we will build our reccomendation system

anime.head()
```

```python
# Checking for missing data 
anime.isnull().sum()
```

```python
anime.dropna(inplace=True) # dropping rows w/ missing vals 
```

```python
# Checking for duplicate data
anime.duplicated().sum()
```

```python
anime = anime.drop_duplicates(keep='first') # keep first duplicate row
```

```python
anime.iloc[0].genre # The way genre is formatted is ok, we don't have to change anything
```

```python
anime['synopsis'][0]
```

```python
anime['synopsis'] = anime['synopsis'].apply(lambda x: x.split()) # Turning each word into a list 
```

```python
anime.head()
```

```python
anime['tags'] = anime['synopsis'] + anime['genre'] # adding a new tags col to df 
```

```python
anime.head()
```

```python
anime.iloc[0].tags
```

```python
new_df = anime[['uid', 'title', 'tags']] # made a new df for ease of use 
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x)) # the tags were prev a list now converted into a string
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower()) # Converting tags to lower case
```
